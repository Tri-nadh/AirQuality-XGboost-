```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('AirQualityUCI.csv')
data.info()
```

```python
print(data.head(),'\n',data.tail())
```

```python
col = data.columns
print(col)
```

```python
for i in col:
    for j in range(data.shape[0]):
        if data[i].iloc[j]== -200:
            data[i].iloc[j] = np.nan
```

```python
data
```

```python
data.isna().sum()
```

```python
df = data.drop(['NMHC'], axis = 1)
df = df.fillna(df.groupby('Date').mean())
df = df.fillna(method = 'ffill')
df
```

```python
df.isna().sum()
```

```python
daysort = df.groupby('Date').mean()
daysort
```

```python
daysort = daysort.reset_index()
daysort
```

```python
train_set = daysort[daysort['index']<= 311]
test_set = daysort[daysort['index']>=311]
```

```python
train_set
```

```python
test_set
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
train_features = train_set.drop(['RH', 'AH'], axis = 1)
train_target_AH = train_set['AH']
train_target_RH = train_set['RH']

test_features = test_set.drop(['RH', 'AH'], axis = 1)
test_target_AH = test_set['AH']
test_target_RH = test_set['RH']
```

```python
from xgboost import XGBRegressor
```

```python
model_1 = XGBRegressor(n_estimators = 600, learning_rate = 0.01)
model_1.fit(train_features,
          train_target_AH,
         eval_set = [(train_features, train_target_AH), (test_features, test_target_AH)],
         eval_metric = 'mae')
```

```python
preds = model_1.predict(test_features)
```

```python
fig = plt.figure(figsize = (20, 9), dpi = 300)
plt.plot(test_target_AH)
```

```python
test_set['Predictions_AH'] = preds
```

```python
fig = plt.figure(figsize = (20, 9))
plt.plot(test_target_AH, label = 'Actual')
plt.plot(test_set['Predictions_AH'], color = 'red', label = 'Predicted')
plt.grid()
plt.legend(loc="upper left")
```
